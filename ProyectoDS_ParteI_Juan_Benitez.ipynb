##Abstrac:
El conjunto de datos "Rendimiento académico y NPS" proporciona información sobre:

**Rendimiento Académico:** El archivo proporciona información sobre el rendimiento académico de los estudiantes a través del número de materias aprobadas, materias regulares y los promedios de calificaciones con y sin aplazos.

**Modalidad de Estudio:** Los datos incluyen modalidades de estudio a distancia, lo que puede ser útil para analizar el impacto de diferentes modalidades en el rendimiento académico y el NPS.

**Datos Demográficos:** La información incluye el género y la provincia de residencia de los estudiantes, lo que permite realizar análisis demográficos.

**Net Promoter Score (NPS):** El NPS es un indicador clave de la satisfacción del estudiante con la institución. Este dato puede ser correlacionado con el rendimiento académico y otros factores para identificar áreas de mejora en la institución.

##Resumen de los datos del archivo:

 4.FECHA_INSCRIPCION: Fecha de la última inscripción del estudiante.

 5.MODALIDAD_DES: Modalidad de estudio (por ejemplo, Distancia - ED, Distancia - MTD).

 6.MATERIAS APROBADAS CARRERA: Número de materias aprobadas en la carrera.

 7.MATERIAS REGULARES: Número de materias regulares en las que el estudiante está inscrito.

 8.PROMEDIO CON APLAZO: Promedio de calificaciones incluyendo aplazos.

 9.PROMEDIO SIN APLAZO: Promedio de calificaciones sin incluir aplazos.

 10.GENERO: Género del estudiante (Femenino, Masculino).

 11.COHORTE: Año de ingreso del estudiante.

 12.PERIODO: Período académico (por ejemplo, 2-2023).

 13.SUBPERIODO: Subperíodo académico (por ejemplo, A-2-2023).

 14.INGR / REINSC: Estado de ingreso o reinscripción del estudiante.

 15.FECHA_PAGO_MATRICULA: Fecha de pago de la matrícula.

 16.PROVINCIA_ALUMNO: Provincia de residencia del estudiante.

 17.INSTITUCION_SECUNDARIO: Institución de educación secundaria del estudiante.

 18.ESTADO: Estado de la matrícula (por ejemplo, MATRICULA PAGA).

 19.ULTIMA_MATRICULA: Último período de matrícula.

 20.ULTIMO_ARANCEL: Último arancel pagado.

##Gráficos univariados

Para comenzar el análisis vamos a comenzar crando un gráfico de Histograma que permita conocer la distribución de los promedios con aplazo y promedios sin aplazo.



```python
sns.set_style('whitegrid')

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 4))


sns.histplot(df_new['PROMEDIO CON APLAZO'], kde=True, color='royalblue', ax=ax1, alpha=0.7)
ax1.set_title('Promedio con Aplazo', fontsize=14, fontweight='bold')
ax1.set_xlabel('Promedio', fontsize=12)
ax1.set_ylabel('Q de estudiantes', fontsize=12)
ax1.set_ylim(0, 250)

sns.histplot(df_new['PROMEDIO SIN APLAZO'], kde=True, color='seagreen', ax=ax2, alpha=0.7)
ax2.set_title('Promedio sin Aplazo', fontsize=14, fontweight='bold')
ax2.set_xlabel('Promedio', fontsize=12)
ax2.set_ylabel('Q de estudiantes', fontsize=12)
ax2.set_ylim(0, 250)

# Ajustes adicionales de la figura
fig.suptitle('Distribución de Promedios', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.subplots_adjust(top=0.88)


plt.show()
```

### Interpretación Histogramas

* Se crearon dos gráficos para represetar los promedios con y sin aplazo.
* Los gráficos tienen un comportamiento similar uno con el otro.
* Podemos observar que el gráfico tiene un volumnen de alumnos que se encuentran con promedio 0. Esto se puede deber a que son almnos que aún no tienen materias arpobadas.
* Ambos histogramas tienen un sesgo hacia la izquierda

```python

colors = ['#FF5733', '#FF5733', '#FF5733', '#FF5733', '#FF5733', '#FF5733', '#FF5733', '#F4D03F', '#F4D03F', '#52BE80', '#52BE80']


counts = df_new['NPS'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
bars = plt.bar(counts.index.astype(str), counts.values, color=colors)


for bar, count in zip(bars, counts):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, count, ha='center', va='bottom', fontsize=10)

plt.title('Distribución del NPS por Puntaje', fontsize=16, fontweight='bold')
plt.xlabel('Puntaje de NPS', fontsize=14)
plt.ylabel('Frecuencia', fontsize=14)

plt.show()
```

###Interpretación gráfico de barra
* El gráfico de barra muestra la distribución del NPS en la población de alumnos.
* Se observa que hay mayor frecuencia en los alumnos que puntúan 9 - 10 (promotores de la universidad)

##Gráficos bivariados


```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Gráfico 1: Promedio con Aplazo vs NPS
sns.scatterplot(x='PROMEDIO CON APLAZO', y='NPS', data=df_new, ax=ax1)
sns.regplot(x='PROMEDIO CON APLAZO', y='NPS', data=df_new, ax=ax1, scatter=False, color='#D98880')
ax1.set_title('Relación entre Promedio con Aplazo y NPS', fontsize=16, fontweight='bold')
ax1.set_xlabel('Promedio con Aplazo', fontsize=14)
ax1.set_ylabel('NPS', fontsize=14)

for spine in ax1.spines.values():
    spine.set_edgecolor('#283747')

# Gráfico 2: Promedio sin Aplazo vs NPS
sns.scatterplot(x='PROMEDIO SIN APLAZO', y='NPS', data=df_new, ax=ax2)
sns.regplot(x='PROMEDIO SIN APLAZO', y='NPS', data=df_new, ax=ax2, scatter=False, color='#D98880')
ax2.set_title('Relación entre Promedio sin Aplazo y NPS', fontsize=16, fontweight='bold')
ax2.set_xlabel('Promedio sin Aplazo', fontsize=14)
ax2.set_ylabel('NPS', fontsize=14)

for spine in ax2.spines.values():
    spine.set_edgecolor('#283747')

plt.tight_layout()
plt.show()
```

###Interpretación de los gráficos de dispersión:
* En los gráficos de dispersión que muestran la relación entre las variables de 'Promedio con Aplazo', 'Promedio sin Aplazo' y 'NPS', no se observa una tendencia clara o evidente entre las variables. Esto sugiere que no hay una relación visual obvia donde cambios en una variable estén correlacionados consistentemente con cambios en otra variable.
* Para una evaluación más precisa de las relaciones entre las variables, se decide utilizar una matriz de correlación. Esta herramienta calcula los coeficientes de correlación entre pares de variables, proporcionando un valor numérico que indica la fuerza y la dirección de la asociación lineal entre ellas. Los valores de correlación pueden oscilar entre -1 y 1, donde:

1. Correlación positiva cercana a 1 indica una relación directa (a medida que una variable aumenta, la otra tiende a aumentar).

2. Correlación negativa cercana a -1 indica una relación inversa (a medida que una variable aumenta, la otra tiende a disminuir).

3. Correlación cercana a 0 indica una falta de relación lineal entre las variables.

```python
correlation_matrix = df_new[['PROMEDIO CON APLAZO', 'PROMEDIO SIN APLAZO', 'NPS']].corr()

plt.figure(figsize=(8, 6))

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt='.2f', linewidths=.5)

plt.title('Matriz de Correlación', fontsize=16, fontweight='bold')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()
plt.show()
```

###Interpretación del mapa de correlación

* En el mapa de correlación se puede evidenciar que hay falta de correlación lineal entre el NPS y los promedios.

###Para continuar el análisis se decidió agrupar las modadlidades presenciales y a distancia para comparar su distribución del NPS.
* Se realizarán graficos de violín y boxplot para su comparación

```python
mapping = {
    'Distancia - ED': 'Distancia',
    'Distancia - MTD': 'Distancia',
    'Presencial HOME': 'Presencial',
    'Presencial HOME  Río Cuarto': 'Presencial',
    'PRESENCIAL.': 'Presencial',
    'PRESENCIAL. RIO IV': 'Presencial'
}

df_new['MODALIDAD_AGRUPADA'] = df_new['MODALIDAD_DES'].map(mapping)

print(df_new)
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='MODALIDAD_AGRUPADA', y='NPS', data=df_new)
plt.title('Comparación del NPS entre Modalidades')
plt.xlabel('Modalidad Agrupada')
plt.ylabel('NPS')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='MODALIDAD_AGRUPADA', y='NPS', data=df_new)
plt.title('Comparación del NPS entre Modalidades')
plt.xlabel('Modalidad Agrupada')
plt.ylabel('NPS')
plt.show()
```

###Interpretación de Gráficos:
Gráfico de Violín y Boxplot:

* Distribución del NPS por Modalidad: Los gráficos de violín y boxplot muestran la distribución de los puntajes de NPS para las modalidades de estudio a distancia y presencial.
 * Modalidad a Distancia: Se observa que la mediana del NPS para la modalidad a distancia es más alta en comparación con la modalidad presencial. Esto indica que, en promedio, los estudiantes que estudian a distancia tienden a tener un mayor nivel de satisfacción.
 * Modalidad Presencial: Aunque también hay estudiantes satisfechos en la modalidad presencial, la mediana del NPS es más baja y hay una mayor dispersión en los puntajes, lo que sugiere una variabilidad mayor en las percepciones de satisfacción entre los estudiantes presenciales.
